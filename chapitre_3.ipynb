## üô§ Chapitre 3 : l'inventaire photographiques d'Eug√®ne Atget

Nous voici munis des biblioth√®ques SPARQLWrapper et RDFlib d√©couvertes dans le chapitre 2, et grace au chapitre 1 d'une bonne compr√©hension du mod√®le de donn√©es de data.bnf.fr.
Il est temps de mettre tout cela √† profit pour extraire du graphe g√©n√©ral de la BnF **toutes les m√©tadonn√©es utiles √† la description du corpus des photographies d'Eug√®ne Atget**.

### Ce [ark:/12148/cb11889340k](https://data.bnf.fr/ark:/12148/cb11889340k), quel relators:pht prolifique

Pour cr√©er un graphe sur les photographies d'Atget, il faut d√©j√† en conna√Ætre la liste. Sur [sa notice d'auteur](https://data.bnf.fr/ark:/12148/cb11889340k) sont recens√©es plus de 4 200 de ses oeuvres iconographiques, impossible d'indexer cela √† la main !
Heureusement nous avons toutes les armes utiles en main üòé

Tout comme [data.bnf.fr/](data.bnf.fr) est la version "pour humain" du point d'acc√®s SPARQL, la [notice d'auteur](https://data.bnf.fr/ark:/12148/cb11889340k) est une vue HTML lisible qui pr√©sente les ressources RDF li√©es √† l'entit√© du graphe repr√©sentant le personnage historique "Eug√®ne Atget". Si l'on interroge cette entit√© avec SPARQL, on peut donc acc√©der √† toutes ses ressources.

> üèóÔ∏è ‚ñ∂Ô∏è √Ä l'aide de SPARQLWrapper et RDFlib, compl√©tez la cellule ci-dessous pour extraite tous les triplets dont l'entit√© `Eug√®ne Atget` est le sujet.

‚ö†Ô∏è **Attention** : Exactement comme pour l'oeuvre photographique du *Cabaret du Soleil d'Or*, l'URI `http://data.bnf.fr/ark:/12148/cb11889340k` r√©f√©rence la repr√©sentation de la **notice d'auteur** dans le graphe, pas l'auteur lui-m√™me en tant que personne. Pour cela, il faut appliquer la m√™me logique que vue dans le chapitre 1 pour d√©r√©f√©rencer la **manifestation** correspondante (mod√®le WEMI), qui repr√©sente la personne elle-m√™me en tant qu'entit√© physique.

```python
from rdflib import Graph
from SPARQLWrapper import SPARQLWrapper, JSONLD

endpoint = SPARQLWrapper('https://data.bnf.fr/sparql') # appel de la m√©thode SQLWRAPPER pour indiquer la source

query = """DESCRIBE
                    <http://data.bnf.fr/ark:/12148/cb11889340k#about>""" # query SPARQL. Les 3 guillemets indiquent √† la machine qu'il ne s'agit pas de code Python

endpoint.setQuery(query) # appel de la requ√™te query

endpoint.setReturnFormat(JSONLD) # indication du format de sortie

g: Graph = endpoint.queryAndConvert() # cr√©ation d'un graph 'g'

print(g)

#¬†V√©rifications
assert isinstance(g, Graph), "Le r√©sultat n'est pas un graphe RDFlib"
assert len(g)  == 16_775, "On attend un graphe de 16 775 triplets"
```

Les triplets obtenus d√©crivent diverses choses, en grand nombre, et utilisent pour cela des propri√©t√©s diff√©rentes.

‚ÑπÔ∏è Pour rappel, un triplet RDF est de la forme (sujet, propri√©t√©, objet).

> ‚ñ∂Ô∏è Grace √† RDFlib, on peut facilement comptabiliser ces propri√©t√©s. Ex√©cutez la cellule suivante puis cliquez sur l'URI des propri√©t√©s pour ouvrir la page Web de leurs d√©finitions, **jusqu'√† trouver une propri√©t√© qui lie Atget √† ses oeuvres**.

```python
#¬†üêç Astuce Python : l'objet Counter de la biblioth√®que standard Python permet de compter facilement des √©l√©ments dans une liste sans r√©√©inventer la roue.
from collections import Counter
properties = [property for subject, property, object in g] #¬†On peut directement it√©rer sur les triplets d'un graphe RDFlib ! Ici pour extraire toutes les propri√©t√©s.
property_counts = Counter(properties)
for prop, count in property_counts.most_common(5):
    print(f"{prop}\t{count} occurrences") 
```

Vous l'avez ? üôå  Ne reste plus qu'√† interroger notre petit graphe local sur Atget pour r√©cup√©rer la liste de toutes ses oeuvres !

La mani√®re la plus simple et propre consiste √† filtrer le graphe `g` √† l'aide de la m√©thode `triples()` qui accepte un "patron" de triplet sous la forme d'un tuple `(sujet, propri√©t√©, objet)` et qui retourne tous les triplets qui correspondent √† ce patron. Allons-y pas √† pas.

> üé¨ D'abord, il faut d√©finir le patron de triplet. Pour cela, rien de tel que regarder un extrait du graphe. Dans le graphe qui s'affiche en ex√©cutant la cellule-ci dessous, rep√©rez :
> - les resources correspondant aux photographies, li√©s par la propri√©t√© recherch√©e

> üèóÔ∏è ‚ñ∂Ô∏è Compl√©tez la d√©finition du patron permettant de **matcher tous les triplets entre Atget et ses photographies**, sachant qu'un patron est un tuple (S, P, O) o√π S, P et O peuvent √™tre :
> 1. None : √©quivalent de "*" dans une expression r√©guli√®re. Matche n'importe quelle valeur.
> 2. Un URI : ne matche que les triplets ayant le m√™me URI.
> 3. Un lit√©ral : ne matche que les triplets ayant la m√™me valeur.

‚ö†Ô∏è Pour que RDFlib reconnaisse une URI il ne faut pas directement passer une cha√Æne de caract√®re mais un objet URIRef. Par exemple, `URIRef("http://data.bnf.fr/ark:/12148/cb11889340k")`.

```python
from rdflib import URIRef

#üèóÔ∏è Compl√©ter le patron pour filtrer les triplets et r√©cup√©rer la liste des photographies de Atget ! 
pattern  = (
    None, #¬†Sujet (en r√©alit√© la liste de photos. None = *)
    URIRef("http://id.loc.gov/vocabulary/relators/pht"), #¬†Propri√©t√© (pr√©dicat)
    URIRef("http://data.bnf.fr/ark:/12148/cb11889340k#about") #¬†Objet (en r√©alit√© le sujet, Eug√®ne Adget)
)

triplets = g.triples(pattern) # ajout √† 'g' des triplets filtr√©s
triplets = list(triplets) # indication que les triplets sont une liste
assert len(triplets) == 4216, f"On attend 4216 triplets pour les photographies d'Atget, mais le patron matche {len(triplets)}."

for s, p, o in triplets:
    print(f"sujet:{s}\tpropri√©t√©:{p}\tobjet={o}")
```

Cr√©ons maintenant un second graphe qui contiendra uniquement l'inventaire des photographies d'Atget, que nous pourrons sauvegarder sur le disque pour l'utiliser comme index des photographies par la suite.

> üèóÔ∏è ‚ñ∂Ô∏è On peut cr√©er un nouvel objet `Graph` vide, puis lui ajouter tr√®s simplement un triplet avec la m√©thode `add(triplet)`. Compl√©tez la cellule suivante pour cr√©er un graphe contenant uniquement les triplets filtr√©s. 

```python
#¬†On cr√©e un nouveau graphe...
atget_photos: Graph = Graph() # cr√©ation du graph vide

#¬†...et on y ajoute les triplets filtr√©s.
for triplet in triplets:
    atget_photos.add(triplet) # ajout des valeurs de la liste au graph
    
assert len(atget_photos) == 4216, f"On attend un graphe de 4216 triplets"
```

Une touche finale : sauvegarder le graphe sur le disque !  C'est encore une fois tr√®s simple, avec la m√©thode `serialize()` du graphe üòä

> üèóÔ∏è ‚ñ∂Ô∏è Compl√©tez la cellule suivante pour sauvegarder le graphe `atget_photos` au format Turtle dans un fichier "g_photos.ttl" du r√©pertoire courant.
